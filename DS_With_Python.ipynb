```python
# read the tables from sql, join thems and retrive it as a data from
df = pd.read_sql(
    "SELECT Name,Unit,Date,FK_Customer_MD_ID,FK_Sale_ID,FK_Product_MD_ID,Amount,TotalPrice,Discount FROM\
    (SELECT Name,P.ID,FK_Category_MD_ID,Unit FROM Category_MD as C join Product_MD as P on C.ID = P.FK_Category_MD_ID) AS t1\
    INNER JOIN\
    (SELECT Date,FK_Customer_MD_ID,SD.ID,FK_Sale_ID,FK_Product_MD_ID,Amount,TotalPrice,Discount FROM Sale as S join Sale_Detail as SD on S.ID = SD.FK_Sale_ID) AS t2\
    ON t1.ID = t2.FK_Product_MD_ID ",con=connection)
```

```python
df1 = df.copy()
```

#### Exteract Year, Month and Day each one as a column

```python
y = []
m = []
d = []
for x in df1["SaleDate"]:
    year, month, day = x.split("/")
    y.append(int(year))
    m.append(int(month))
    d.append(int(day))
df1["Year"] = y
df1["Month"] = m
df1["day"] = d
df1.head()
```

```python
#drop rows with Amount == 0
df1.drop(df1[df1.Amount == 0].index,inplace=True)
```

```python
# Divide TotalPrice on Amount to calculate UnitPrice
df1["UnitPrice"] = df1["TotalPrice"] / df1["Amount"]
```

```python
# Filter 20019 Product ID
df2 = df1.loc[df1.ProductID == 20019]
```

```python
df2.info()
```

```python
# Group by year and month and average on unitprice
df3 = df2.groupby(['Year','Month']).UnitPrice.agg(["mean"])
```

```python
# when we use group by it returns datafram with multiindex
type(df3.index)
```

```python
# we reset it here to have a deafult datafram
df3 = df3.reset_index()
```

```python
df3.info()
```

```python
# Name Persian Month for number of each month
conditions = df3.Month == 1, df3.Month == 2, df3.Month == 3, df3.Month == 4,\
df3.Month == 5, df3.Month == 6, df3.Month == 7, df3.Month == 8,\
df3.Month == 9, df3.Month == 10, df3.Month == 11, df3.Month == 12

outputs = ['Farvardin', 'Ordibehest', 'Khordad', "Tir", "Mordad", "Shahrivar", "Mehr",
    "Aban", "Azar", "Day", "Bahman", "Esfand"]

df3['Month(Name)'] = np.select(conditions, outputs)
```

```python
# Concatenate Year and Month name
df3["YearMonth"] = df3["Year"].astype(str) + "-" + df3["Month(Name)"]
```
